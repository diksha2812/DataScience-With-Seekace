```python
# read csv named "titanic.csv"
import pandas as pd
df=pd.read_csv("titanic.csv")
df
```

```python
df.info()
```

```python
# Total number of missing values in each column of dataframe "df"
df.isnull().sum()
```

```python
# Drop column i.e, Cabin from the dataframe "df"
df.drop(["Cabin","Name","PassengerId"], axis=1, inplace=True)
df
```

```python
# Value count of column "Embarked"
df.Embarked.value_counts()
```

```python
# Replacing null or empty values in column "Embarked" with S 
df["Embarked"].fillna("S", inplace = True)
df
```

```python
# Checking weather the null values replaced or not 
df.Embarked.value_counts()
```

```python
# finding mean of column "Age"
df.Age.mean()
```

```python
# Replacing null or empty values in column "Age" with mean value
df["Age"].fillna("29.6", inplace = True)
df
```

```python
# Again checking the missing values
df.isnull().sum()
```

```python
# There is no duplicate value in our Dataframe "df"
df.duplicated().sum()
```

```python
# Droping the duplicates values from the dataframe "df"
df.drop_duplicates(inplace = True)
df
```

```python
df.describe()
```

```python
# Correlation 
import matplotlib.pyplot as plt
import seaborn as sns
colormap = plt.cm.cubehelix_r
plt.figure(figsize=(16,12))

plt.title('correlation of numeric features', y=1.05, size=15)
sns.heatmap(df.corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
le_Embarked = LabelEncoder()
```

```python
df["Sex_n"] = le_Sex.fit_transform(df['Sex'])
df["Embarked_n"] = le_Embarked.fit_transform(df['Embarked']) 
df
```

```python
df.drop(["Sex","Embarked"],axis=1, inplace=True)
df
```

```python
df.drop(["Ticket"],axis=1, inplace=True)
df
```

```python
inputs = df.drop(["Survived"],axis=1)
target = df["Survived"]
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
model.fit(inputs, target)
```

```python
model.score(inputs, target)
```

```python
from sklearn.metrics import classification_report

predictions=model.predict(X_test)
print(classification_report(y_test,predictions))
```
