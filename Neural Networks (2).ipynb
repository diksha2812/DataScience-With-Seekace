# Implementation of OR GATE

```python
import numpy as np # to get numpy function 
```

Input values:

```python
# Assign Input values:
feature_set = np.array([[0,0],[0,1],[1,0],[1,1]])
print(feature_set.shape)
feature_set
```

Target Output

```python
# Target Output:
target_output = np.array([[0,1,1,1]])
target_output = target_output.reshape(4,1)
print(target_output.shape)
target_output
```

```python
# Assign the weights

weights=np.array([[0.1],[0.2]])
print(weights.shape) # Printing shape of Weights
weights # printing Weights
```

```python
# Adding Bias Values and Assigning a Learning Rate :

#bias weight
bias=0.3

# Learning Rate
lr=0.05
```

```python
#Applying a Sigmoid Function:

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# Derivative of sigmoid function:

def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
# epoch is simply iterations 
# The main logic for predicting output and updating the weight values:

for epoch in range(10000):
    inputs = feature_set

    # feedforward input
    in_o = np.dot(feature_set, weights) + bias

    #feedforward output
    out_o = sigmoid(in_o)


    # backpropagation calculating error
    error = out_o - target_output

    # applying formula
    print(error.sum())

    # Calculating derivative
    derror_douto = error
    douto_dino = sigmoid_der(out_o)

    # Multiplying individual derivatives
    deriv = derror_douto * douto_dino

    # Multiplying with the 3rd individual derivative
    # finding the transpose of input features
    inputs = feature_set.T
    deriv_final=np.dot(inputs, deriv)
    
    #updating the weight values
    weights -= lr * deriv_final

    # updating the bias weight value
    for i in deriv:
        bias -= lr * i
```

```python
# Pritng Weights and bias
print(weights)
print(bias)
```

```python
# Predictiong for (0,0)

#taking input
pt=np.array([0,0])

# 1st step
result1= np.dot(pt,weights)+bias

# 2nd step
result2=sigmoid(result1)
    
#print result
print(result2)
```

```python
# Predictiong for (0,1)

#taking input
pt=np.array([0,1])

# 1st step
result1= np.dot(pt,weights)+bias

# 2nd step
result2=sigmoid(result1)
    
#print result
print(result2)
```

```python
# Predictiong for (1,0)

#taking input
pt=np.array([1,0])

# 1st step
result1= np.dot(pt,weights)+bias

# 2nd step
result2=sigmoid(result1)
    
#print result
print(result2)
```

```python
# Predictiong for (1,1)

#taking input
pt=np.array([1,1])

# 1st step
result1= np.dot(pt,weights)+bias

# 2nd step
result2=sigmoid(result1)
    
#print result
print(result2)
```
