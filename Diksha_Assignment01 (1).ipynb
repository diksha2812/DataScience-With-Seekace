```python
# Changing the type of Column 'Age' from string to int in dataframe df
df['Age']=df['Age'].astype(str).astype(int)
```

```python
"""# Question12.  Create a new column Age_Group as per the following conditions 
Age    Age_Group
0-40    1
41-60   2
61-80   3
81-100  4"""
import numpy as np
conditions = [
    (df['Age'] > 0) & (df['Age'] <= 40),
    (df['Age'] > 40) & (df['Age'] <= 60),
    (df['Age'] > 60) & (df['Age'] <= 80),
    (df['Age'] > 80)  
    ]
#list of the values we want to assign for each condition
values = [1, 2, 3, 4]
# creating a new column and use np.select to assign values to it using our lists as arguments
df['Age_Group'] = np.select(conditions, values)

# display updated DataFrame
print(df.to_string())
```
